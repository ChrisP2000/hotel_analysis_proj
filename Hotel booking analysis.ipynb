## DATA ANALYSIS and Visualization

```python
#counting the percetage of cancelled and non cancelled 
cancelled_perc = booking_df['is_canceled'].value_counts(normalize=True)
cancelled_perc

#now we will visualize it using bar graph 
plt.figure(figsize= (5,4))
plt.title('Reservation Status ')
plt.bar(['cancelled','not cancelled'], booking_df['is_canceled'].value_counts(), width=0.7)
```

```python
plt.figure(figsize =(8,4))
g= sns.countplot(x= 'hotel', hue ='is_canceled', data= booking_df, palette= 'Blues')
legend_labels ,_= g. get_legend_handles_labels()
new_labels= ['not cancelled','cancelled']
g.legend(labels= new_legend,bbox_to_anchor=(1,1))    
plt.title("Reservation counts, group by hotels", size= 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
# filter out the hotels 
# And find the perc of canc and non canc in each hotel 
resort_hotel = booking_df[booking_df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python

city_hotel = booking_df[booking_df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize= True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
print(city_hotel)
```

```python
plt.figure(figsize=(20,8))


plt.title('Average Daily Rate in City and Resort Hotels', size=20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'resort_hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'city_hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# maybe can check the dopsit type with if the person will refund or not 
booking_df['deposit_type'].value_counts( normalize= True)

non_refund= booking_df[booking_df['deposit_type']=='Non Refund']
is_cancel = non_refund.groupby('is_canceled')[['deposit_type']].count()

print(is_cancel)
plt.figure(figsize=(5, 4))
plt.bar(is_cancel.index, is_cancel['deposit_type'], tick_label=['Not Canceled', 'Canceled'])
plt.xlabel('Cancellation Status')
plt.ylabel('Count of Non-Refundable Bookings')
plt.title('Non-Refundable Bookings Cancellation Status')
plt.show()
```

```python
#finding out the canceled reservation per month 
booking_df['month']=booking_df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
graph = sns.countplot(x='month', hue='is_canceled', data= booking_df, palette = 'Set2')
legend_lebels,_ = graph.get_legend_handles_labels()



total = booking_df.groupby('month').size().values
for p in graph.patches:
    height = p.get_height()
    total_for_month = total[int(p.get_x())]
    percentage = (height / total_for_month) * 100
    graph.text(p.get_x() + p.get_width() / 2., height, '{:.1f}%'.format(percentage), ha='center', va='bottom')
    
plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
legend_labels, _ = graph.get_legend_handles_labels()
plt.legend(['Not Canceled', 'Canceled'], title='Reservation Status')
plt.show()
```
